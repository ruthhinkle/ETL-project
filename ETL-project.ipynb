```python
import pandas as pd
from sqlalchemy import create_engine
import datetime
from config import password
```

```python
imdb_file = "Resources/series_data.csv"
imdb_df = pd.read_csv(imdb_file)
imdb_df.head()
```

```python
netflix_file = "Resources/tv_shows_data.csv"
netflix_df = pd.read_csv(netflix_file)
netflix_df.head()
```

```python
# FROM IMDB: genre, imdb rating, overview, 
# FROM NETFLIX: No_of_Seasons, title, Premiere, No_of_Episodes
```

```python
# Pull specificied columns from each dataframe into a new dataframe
new_netflix_df = netflix_df[['Title', 'Premiere', 'No_of_Seasons', 'No_of_Episodes']]
new_netflix_df.head()
```

```python
new_imdb_df = imdb_df[['Series_Title','Genre', 'IMDB_Rating', 'Overview']]
new_imdb_df.head()
```

```python
# Find out type for all column in new_netflix_df
new_netflix_df.dtypes
```

```python
# Convert Premiere object to a date
# Remove dashes from date in "Premiere" column
new_netflix_df['Premiere'] = new_netflix_df['Premiere'].str.replace('-', '')
```

```python
# display new_netflix_df with no space in date
new_netflix_df
```

```python
# Format date from string in series to datetime format
# Found duplicate column header rows & one-off "awaiting release" row, removed in the most efficient way by deleting from original csv and re-uploading into pandas. 
new_netflix_df['Premiere'] =  pd.to_datetime(new_netflix_df['Premiere'], format='%d%b%y')
```

```python
# display corrected df
new_netflix_df
```

```python
# new_netflix_df.set_index("id", inplace=True)
```

```python
# Create database connection
connection_string = f'postgres:{password}@localhost:5432/shows_db'
engine = create_engine(f'postgresql://{connection_string}')  
```

```python
# Confirm tables
engine.table_names()
```

```python
new_netflix_df = new_netflix_df.rename(columns={"Title": "title",
                                                  "Premiere": "premiere",
                                                  "No_of_Seasons": "no_of_seasons",
                                               "No_of_Episodes": "no_of_episodes"})
new_netflix_df
```

```python
# Rename columns for imdb df
new_imdb_df = new_imdb_df.rename(columns={"Series_Title": "series_title",
                                                  "Genre": "genre",
                                                  "IMDB_Rating": "imdb_rating",
                                               "Overview": "overview"})
new_imdb_df
```
